```python
data_model = df_model['adjacent_overlap_binary_noun_para'].values
df_model['adjacent_overlap_binary_noun_para'].describe()
```

```python
data_texttiling = df_texttiling['adjacent_overlap_binary_noun_para'].values
df_texttiling['adjacent_overlap_binary_noun_para'].describe()
```

```python
data_random = df_random['adjacent_overlap_binary_noun_para'].values
df_random['adjacent_overlap_binary_noun_para'].describe()
```

```python
# dictionary for colours
names = ['Model', 'TextTiling', 'Random']
cdict = dict(zip(names, ['blue','red', 'green']))

plt.figure(figsize=(14,8))
for i, data in enumerate([data_model, data_texttiling, data_random]):
    mean = np.mean(data)
#     plt.axvline(x=mean, ymin=0, ymax=0.9, label=names[i], color=cdict[names[i]])
    n, bins, patches = plt.hist(data, 500, histtype='step', normed=True, color=cdict[names[i]])
# plt.axis([-5, 5, 0, 1.2])
plt.title("Timing at turn transitions")
plt.xlabel("Transition time (s)")
plt.ylabel("Relative Frequency")
plt.legend()
```
